```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
```

```python
driver=webdriver.Chrome()
driver.maximize_window()
```

```python
url="https://www.cars24.com/buy-used-car?sort=bestmatch&serveWarrantyCount=true&listingSource=ViewAllCars&storeCityId=1"
elements=driver.get(url)
for k in range(20):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)
    element=driver.find_elements(By.CLASS_NAME,"_2YB7p")
```

```python
cars_data =[]
for cars in element:
    try:
        brand_name = cars.find_element(By.CLASS_NAME,"_11dVb").text
    except:
        brand_name = "NA"
    try:
        fuel_type = cars.find_element(By.CLASS_NAME,"_3J2G-").text
    except:
        fuel_type = "NA"
    try:
        price = cars.find_element(By.CLASS_NAME,"_3RL-I").text
    except:
        price = "NA"
    try:
        emi = cars.find_element(By.CLASS_NAME,"-EPGU").text
    except:
        emi = "NA"
    try:
        location=cars.find_element(By.CLASS_NAME,"_3dGMY").text
    except:
        location = "NA"
    
    cars_data.append({'name':brand_name,'fuel' : fuel_type,'prices' : price, 'emis' : emi, 'Locations' : location})
```
